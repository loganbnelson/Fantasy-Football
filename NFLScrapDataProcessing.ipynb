```python
QBefficiency = QBStats.query('Season > 2019')

QBefficiency = QBefficiency.groupby('Player').mean().sort_values('Fantasy Points', ascending=False).drop(columns = ['Game', 'Season', 'Week', 'SeasonWeek'])

QBefficiency['Completion Rate'] = QBefficiency['Passing Completions'] / QBefficiency['Passing Attempts']

QBefficiency['Passing Yards per Completion'] = QBefficiency.loc[:,'Passing Yards'] / QBefficiency.loc[:,'Passing Completions']

QBefficiency['Passing TDs per Completion'] = QBefficiency['Passing TDs'] / QBefficiency['Passing Completions']

QBefficiency['Rushing Yards per Attempt'] = QBefficiency[' Rushing Yards'] / QBefficiency['Rushing Attempts']

QBefficiency['Rushing TDs per Attempt'] = QBefficiency['Rushing TDs'] / QBefficiency['Rushing Attempts']

QBefficiency['INTs per Attempt'] = QBefficiency['Passing INTs'] / QBefficiency['Passing Attempts']

QBefficiency.head()
```

```python
RBefficiency = RBStats.query('Season > 2019')

RBefficiency = RBefficiency.groupby('Player').mean().sort_values('Fantasy Points', ascending=False).drop(columns = ['Game', 'Season', 'Week', 'SeasonWeek'])

RBefficiency['Catch Rate'] = RBefficiency['Reception'] / RBefficiency['Receiving Target']

RBefficiency['Receiving Yards per Completion'] = RBefficiency['Receiving Yards'] / RBefficiency['Reception']

RBefficiency['Receiving TDs per Completion'] = RBefficiency['Receiving TDs'] / RBefficiency['Reception']

RBefficiency['Rushing Yards per Attempt'] = RBefficiency[' Rushing Yards'] / RBefficiency['Rushing Attempts']

RBefficiency['Rushing TDs per Attempt'] = RBefficiency['Rushing TDs'] / RBefficiency['Rushing Attempts']

# RBefficiency['Fumbles per Attempt'] = RBefficiency[''] / RBefficiency['']

RBefficiency.head()
```
