```python
urlstopull = []
# print(urlstopull)

BASE_URL = 'https://www.fftoday.com/stats/playerstats.php?Season='

# https://www.fftoday.com/stats/playerstats.php?Season=2015&GameWeek=13&PosID=30&LeagueID=17
# &order_by=FFPts&sort_order=DESC&cur_page=2

seasons_desired = [*range(2013,2023,1)]
# print(seasons_desired)

gameweekcombos = [*range(1,19,1)]
# print(gameweekcombos)

pages = [*range(0,3,1)]

PosIDs = {'QB':10,'RB':20,'WR':30,'TE':40,'K':80,'DEF':99}
PosIDvalues = PosIDs.values()
# print(PosIDs)

combinedvaluesdict = {}
iteration = seasons_desired[0]

for season in seasons_desired:
    urlstopull.append(BASE_URL+str(season)+'&GameWeek=')
    combinedvaluesdict[season] = {'season':season}

urlstopull2 = []
combinedvalueslist = []

for url in urlstopull:
    for game in gameweekcombos:
        urlstopull2.append(url+str(game)+'&PosID=')
        combinedvalueslist.append([iteration,game])
    iteration = iteration + 1
    
    
urlstopull3 = []
combinedvalueslist2 = []
iteration = 0

for url in urlstopull2:
    for page in pages:
#         urlstopull3.append(url+str(10)+'&LeagueID=17')
        combinedvalueslist2.append([10]+combinedvalueslist[iteration])
    iteration = iteration + 1

urlstopullQB = []
for url in urlstopull2:
    for page in pages:
        urlstopullQB.append(url+str(10)+'&LeagueID=17&order_by=FFPts&sort_order=DESC&cur_page='+str(page))
print(len(urlstopullQB))
print(urlstopullQB[0:10])
    
urlstopullRB = []
for url in urlstopull2:
    for page in pages:
        urlstopullRB.append(url+str(20)+'&LeagueID=17&order_by=FFPts&sort_order=DESC&cur_page='+str(page))
print(len(urlstopullRB))

urlstopullWR = []
for url in urlstopull2:
    for page in pages:
        urlstopullWR.append(url+str(30)+'&LeagueID=17&order_by=FFPts&sort_order=DESC&cur_page='+str(page))
print(len(urlstopullWR))

urlstopullTE = []
for url in urlstopull2:
    for page in pages:
        urlstopullTE.append(url+str(40)+'&LeagueID=17&order_by=FFPts&sort_order=DESC&cur_page='+str(page))
print(len(urlstopullTE))

urlstopullDEF = []
for url in urlstopull2:
    for page in pages:
        urlstopullDEF.append(url+str(99)+'&LeagueID=17&order_by=FFPts&sort_order=DESC&cur_page='+str(page))
print(len(urlstopullDEF))

urlstopullK = []
for url in urlstopull2:
    for page in pages:
        urlstopullK.append(url+str(80)+'&LeagueID=17&order_by=FFPts&sort_order=DESC&cur_page='+str(page))
print(len(urlstopullK))
        
# print(len(combinedvalueslist2))        
# print(len(urlstopull3))
# # note that the combinedvalueslist2 structure is "position", "year/season", "game"
# print(combinedvalueslist2[0:10])
# print(len(combinedvalueslist2))
# print(urlstopull3[0:10])
```
